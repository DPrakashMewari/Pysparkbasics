```python
emp1 = spark.createDataFrame(data=emp_data1,schema=emp_schema)
emp2 = spark.createDataFrame(data=emp_data2,schema=emp_schema)
```

```python
emp = emp1.union(emp2)
```

```python
emp.show()
```

```python
from pyspark.sql.functions import asc,desc,col

# select * from emp order by salary desc;

emp_desc = emp.orderBy(col("salary").asc());
```

```python
emp_desc.show()
```

```python
from pyspark.sql.functions import count,sum
```

```python
emp_agg = emp_desc.groupby("department").agg(count("emp_id").alias("department count"))  # department wise 
```

```python
emp_agg.show()
```

```python
emp_agg1 = emp_desc.groupby("Role").agg(sum("salary").alias("total salary based on role"))  #role 
```

```python
emp_agg1.show()
```

```python
# aggregate 
# SELECT class, AVG(marks) AS avg_marks
# FROM students
# GROUP BY class
# HAVING AVG(marks) > 75;

emp_agg3 = emp_desc.groupby("department").agg(count("emp_id").alias("department_count")).where("department_count >2")                                                                                                  # department wise 
```

```python
emp_agg3.show()
```
